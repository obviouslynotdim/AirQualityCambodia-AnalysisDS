## Fetch Data from OpenAQ API
We collect PM2.5 data from Phnom Penh using OpenAQ.

```python
url = "https://api.openaq.org/v2/measurements"

params = {
    "country": "KH",     
    "city": "Phnom Penh", 
    "parameter": "pm25",  
    "limit": 1000,    
    "sort": "desc"
}

response = requests.get(url, params=params)
data = response.json()

records = data.get("results", [])

if not records:
    print("‚ö†Ô∏è No data returned. Check city name or parameter.")
else:
    df = pd.json_normalize(records)
    print(df.head())
    
# note: not working as of now since no api !!! < nak
```
